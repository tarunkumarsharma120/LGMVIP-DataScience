# <font face='Bahnschrift'><font color=green><center><b> IMPORT TENSORFLOW 2.0

```python
import tensorflow as tf
```

```python
mnist=tf.keras.datasets.mnist
```

# <font face='Bahnschrift'><font color=green><center><b> LOAD MNIST DATASET FROM KERAS DATASETS

```python
(X_train,y_train),(X_test,y_test)=mnist.load_data()
```

```python
import matplotlib.pyplot as plt
```

```python
plt.imshow(X_train[0])
```

```python
X_train[0]
```

# <font face='Bahnschrift'><font color=green><center><b> NORMALIZE THE DATA

```python
X_train=X_train/255
X_test=X_test/255
```

```python
X_train[0]
```

```python
plt.imshow(X_train[0])
```

# <font face='Bahnschrift'><font color=green><center><b> FORMATION OF CNN

```python
from tensorflow.keras import layers,models
import numpy as np
```

```python
X_train[0].shape
```

```python
X_trainshape=np.array(X_train).reshape(-1,28,28,1)
X_trainshape.shape[1:]
```

```python
model=models.Sequential([layers.Conv2D(filters=20,kernel_size=(3,3), activation='relu',input_shape=(28,28,1)),
                  layers.MaxPooling2D((2,2)),

                  layers.Conv2D(filters=40,kernel_size=(3,3), activation='relu'),
                  layers.MaxPooling2D((2,2)),

                  layers.Conv2D(filters=60,kernel_size=(3,3), activation='relu'),
                  layers.MaxPooling2D((2,2)),



                  layers.Flatten(),
                  layers.Dense(60,activation='relu'),
                  layers.Dense(40,activation='relu'),
                  layers.Dense(10,activation='softmax')
                  ])
```

# <font face='Bahnschrift'><font color=green><center><b> MODEL SUMMARY

```python
model.summary()
```

# <font face='Bahnschrift'><font color=green><center><b> MODEL TRAINING

```python
model.compile(optimizer='adam',loss='sparse_categorical_crossentropy',metrics=['accuracy'])
model.fit(X_trainshape,y_train,epochs=15)
```

# <font face='Bahnschrift'><font color=green><center><b> MODEL PREDICTION

```python
from sklearn.metrics import confusion_matrix, classification_report
import numpy as np
y_pred=model.predict(np.array(X_test).reshape(-1,28,28,1))
```

```python
plt.imshow(X_test[0])
```

```python
np.argmax(y_pred[0])
```

```python
plt.imshow(X_test[201])
```

```python
np.argmax(y_pred[201])
```

# <font face='Bahnschrift'><font color=green><center><b> MODEL EVALUATION

```python
model.evaluate(np.array(X_test).reshape(-1,28,28,1),y_test)
```

```python
from sklearn.metrics import confusion_matrix, classification_report
y_predicted_labels = [np.argmax(i) for i in y_pred]
print("Classfication report\n",classification_report(y_test,y_predicted_labels))
import seaborn as sns

sns.heatmap(confusion_matrix(y_test,y_predicted_labels),annot=True)
```
