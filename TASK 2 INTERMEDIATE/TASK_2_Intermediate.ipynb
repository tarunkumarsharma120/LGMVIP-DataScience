### <font face='MV Boli'><font color=green><b>Reading the CSV File.

```python
import sklearn.datasets as datasets
import pandas as pd

iris=pd.read_csv('Iris.csv')

iris.columns
```

### <font face='MV Boli'><font color=green><b> Preprocessing of the data.

```python
from sklearn.preprocessing import LabelEncoder
enc = LabelEncoder()
y = enc.fit_transform(iris.Species)
```

```python
df=pd.DataFrame(iris, columns=['Id', 'SepalLengthCm', 'SepalWidthCm', 'PetalLengthCm', 'PetalWidthCm'])
print(df.head(5))

print(y)
```

### <font face='MV Boli'><font color=green><b>Now let us define the Decision Tree Algorithm

```python
from sklearn.tree import DecisionTreeClassifier
dtree=DecisionTreeClassifier()
dtree.fit(df,y)

print('Decision Tree Classifer Created')
```

### <font face='MV Boli'><font color=green><b>Let us visualize the Decision Tree to understand it better.



```python
from six import StringIO  
from IPython.display import Image  
from sklearn.tree import export_graphviz
import pydotplus

dot_data = StringIO()
export_graphviz(dtree, out_file=dot_data, feature_names=['Id', 'SepalLengthCm', 'SepalWidthCm', 'PetalLengthCm', 'PetalWidthCm'],  
                filled=True, rounded=True,
                special_characters=True)
graph = pydotplus.graph_from_dot_data(dot_data.getvalue())  
Image(graph.create_png())
```
